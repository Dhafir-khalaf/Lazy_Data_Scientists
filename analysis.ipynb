## Importing and implementing Sweetviz library
### There are 3 main functions for creating reports:

* **analyze(...)**: to analyze one data set
* **compare(...)**: to compare two data sets
* **compare_intra(...)**: to compare categories within one column

```python
import sweetviz as sv
my_report = sv.analyze(data)
my_report.show_html() # Default arguments will generate to "SWEETVIZ_REPORT.html" saves it to the same directory
```

```python
my_report.show_notebook() #Here you can show the report inside the notebook
```

## 3-LazyPredict: 

This library allows you to compare different types of models:

*you should install it first to your running environment through:* **pip install lazypredict**

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

df = load_breast_cancer()
X = df.data
y= df.target

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.5,random_state =123)

clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

print(models)
```
